```python
# import machine geometry
import os
import json

from pyemmo.api.json.modelJSON import importMachineGeometry, createMachineGeometryFromSegment
from pyemmo.api.json.importJSON import importExtInfo, getSymFactor
from pyemmo.definitions import ROOT_DIR

# os.path.join(ROOT_DIR, "Results")

geo = os.path.join(ROOT_DIR, "Results/jsonAPI/TTZ_1FE1051-4HF11_TherCom/TTZ_1FE1051-4HF11_TherCom.json")
extInfo = os.path.join(ROOT_DIR, "Results/jsonAPI/TTZ_1FE1051-4HF11_TherCom/simuInfo.json")

if isinstance(geo, str):
    if os.path.isfile(geo):
        # import the segment surface list from the json file
        try:
            with open(geo, encoding="utf-8") as jsonFile:
                machineGeoList = json.load(jsonFile)
                # create dict with surface api (segment) objects from the surface list
                segmentSurfDict = importMachineGeometry(
                    machineGeoList
                )
        except FileNotFoundError as fnfe:
            raise fnfe
        except Exception as exept:
            raise exept
    else:
        raise (FileNotFoundError(f"Given file path {geo} was not a file."))
elif isinstance(geo, dict):
    segmentSurfDict = geo
else:
    raise TypeError(
        f"Geometry file has to be type 'File' or 'dict', not {type(geo)}"
    )

# addition information
if isinstance(extInfo, str):
    if os.path.isfile(extInfo):
        # import the extended information
        extendedInfo = importExtInfo(extInfo)
    else:
        raise (
            FileNotFoundError(f"Given file path {extInfo} was not a file.")
        )
elif isinstance(extInfo, dict):
    extendedInfo = extInfo

symFactor = getSymFactor(extendedInfo)

machineSurfDict = createMachineGeometryFromSegment(segmentSurfDict, symFactor)

machineSurfDict
```
